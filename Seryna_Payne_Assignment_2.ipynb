#2. Data Cleaning and Preprocessing

2.1 Remove the first column or Column 1 from the dataset. Treat the year variable as nominal.

```python
# remove 'Column 1'
df=df.drop(columns=['Column 1'])

# treat the 'year' variable as nominal
df['year']=df['year'].astype('category')
```

2.2 Check for duplicate values and remove them.

```python
# remove duplicate values
df.drop_duplicates(inplace=True)
```

2.3 Check for missing values. If a data record (row) only has a few missing values, replace the missing values with the median of the column feature in that specific Region variable. If most column values in a data record are missing, remove the data record.

```python
# define threshold for "most"
missing_threshold_percent = 0.5

# caluclate number of missing values per row
missing_count_per_row = df.isnull().sum(axis=1)

# calculate the threshold
total_columns = df.shape[1]
missing_threshold_count = total_columns * missing_threshold_percent

# keep rows where missing values is less than threshold
df = df[missing_count_per_row < missing_threshold_count]

# identify columns to impute
columns_to_impute = df.select_dtypes(include=[np.number]).columns.drop('region', errors='ignore')

for col in columns_to_impute:
    # calculate the median for each region for the current column
    region_medians = df.groupby('region')[col].median()

    # fill NA values in the column using the calculated region medians
    df[col] = df[col].fillna(df['region'].map(region_medians))
```

3.3 Highest average price
- Find the highest value in "AveragePrice".
- Report which region it belongs to.

```python
# find the highest value in "AveragePrice"
highest_average_price = df['AveragePrice'].max()
print("Highest Average Price: $", highest_average_price)

# find which region it belongs to
highest_average_price_region = df.loc[df['AveragePrice'] == highest_average_price, 'region'].values
print("Region with Highest Average Price:", highest_average_price_region)

# I used the df.loc function from the Pandas library which found which region value was associated with the highest average price value.
```
